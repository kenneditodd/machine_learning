```python
# Kennedi Todd
# July 30, 2024
# Chapter 18 Exercise

# import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import ensemble
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import GridSearchCV

# read csv
df = pd.read_csv('/Users/kenneditodd/Documents/machine_learning/machine_learning_for_absolute_beginners_third_edition/Melbourne_housing_FULL.csv')
# preview
#df.head(10) 

# scrub data
del df['Address']
del df['Method']
del df['SellerG']
del df['Date']
del df['Postcode']
del df['Lattitude']
del df['Longtitude']
del df['Regionname']
del df['Propertycount']

# remove rows with missing values
df.dropna(axis = 0,        # 0 = row, 1 = column
          how = 'any',     # drop rows if at least one value missing
          subset = None,   # use all columns
          inplace = True)  # update rather than replace

# one-hot encoding
df = pd.get_dummies(df, columns = ['Suburb', 'Type', 'CouncilArea'])

# assign independent and dependent variables
X = df.drop('Price', axis = 1)
y = df['Price']

# split the data 70/30
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, shuffle = True)

# gradient boosting regressor
model = ensemble.GradientBoostingRegressor() 

# grid search
hyperparameters = {
    'n_estimators': [200,300],
    'max_depth': [8,6],
    'min_samples_split': [8,10],
    'min_samples_leaf': [5,6],
    'learning_rate': [0.01,0.02],
    'max_features': [0.8,0.9],
    'loss': ['ls', 'lad', 'huber']
}
grid = GridSearchCV(model, hyperparameters, n_jobs = 4)
grid.fit(X_train, y_train)

# return best params
grid.best_params_

# evaluate results
mae_train = mean_absolute_error(y_train, grid.predict(X_train))
print("Training Set Mean Absolute Error: %.2f" % mae_train)
```

```python
mae_test = mean_absolute_error(y_test, grid.predict(X_test))
print("Test Set Mean Absolute Error: %.2f" % mae_test)
```
