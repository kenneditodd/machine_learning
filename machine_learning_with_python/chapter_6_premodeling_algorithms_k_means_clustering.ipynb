```python
# Kennedi Todd
# August 2, 2024
# Chapter 6: Premodeling Algorithms
# k-means clustering & scree plot

# libraries
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline

# use make_blobs instead of importing a dataset
X, y = make_blobs(n_samples = 300,
                  n_features = 2,
                  centers = 4,
                  cluster_std = 4,
                  random_state = 10)

# plot random data
plt.figure(figsize = (7,5))
plt.scatter(X[:, 0], X[:, 1])
```

```python
# set algorithm
model = KMeans(n_clusters = 4)
model.fit(X)

# predict
model_predict = model.predict(X)
centroids = model.cluster_centers_
print(model.cluster_centers_)
```

```python
# visualize
plt.figure(figsize = (7,5))
plt.scatter(X[:, 0],
            X[:, 1],
            c = model_predict,
            s = 50,
            cmap = 'rainbow')
plt.scatter(centroids[:, 0],
            centroids[:, 1],
            c = 'black',
            s = 200,
            alpha = 1)
```

```python
# optimize k
distortions = []
K = range(1,10)
for k in K:
    model = KMeans(n_clusters = k)
    model.fit(X, y)
    distortions.append(model.inertia_)

# scree plot
plt.figure(figsize = (16,8))
plt.plot(K, distortions)
plt.xlabel('k')
plt.ylabel('Distortion')
plt.show()
```
