```python
# Kennedi Todd
# August 2, 2024
# Chapter 7: Split Validation

# libraries
from sklearn.model_selection import train_test_split
import pandas as pd

# read data
df = pd.read_csv('advertising.csv')
df.head()
```

```python
# assign independent and dependent variables
X = df.drop(['City','Male','Timestamp','Clicked on Ad'], 
            axis = 1, 
            inplace = False)
y = df['Clicked on Ad']

# split validation with training/testing at 70/30
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size = 0.3,
                                                    random_state = 10, # seed number
                                                    shuffle = True)
```
